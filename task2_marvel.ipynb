```python
import matplotlib as mpl
import numpy as np
import matplotlib.pyplot as plt

```

```python
# line plot - A line plot is a type of chart or graph which displays information as a series of data points
x = np.linspace(0,50,100)
y = x * np.linspace(100,150,100)

plt.plot(x,y)
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.xlim(0,50)
plt.ylim(1200, 10000)
```

```python
# creates a 2x2 grid of subplots, plots data in each subplot with different colors, sets titles for each subplot, and adds x-labels and y-labels

fig, axs = plt.subplots(2, 2)
axs[0, 0].plot(x, y)
axs[0, 0].set_title('Axis [0, 0]')
axs[0, 1].plot(x, y, 'tab:orange')
axs[0, 1].set_title('Axis [0, 1]')
axs[1, 0].plot(x, -y, 'tab:green')
axs[1, 0].set_title('Axis [1, 0]')
axs[1, 1].plot(x, -y, 'tab:red')
axs[1, 1].set_title('Axis [1, 1]')

for ax in axs.flat:
    ax.set(xlabel='x-label', ylabel='y-label')
for ax in axs.flat:
    ax.label_outer()

```

```python
 #this plot displays two datasets (y1 and y2) as line plots with different colors and provides a legend to label and distinguish between them
a = [10,20,30,40,50]
y1 = [1, 2, 3, 4, 5]
y2 = [1, 4, 9, 16, 25]
plt.plot(a, y1,color='pink', label ='y1')
plt.plot(a, y2,color='yellow',label ='y2')
plt.legend(loc="lower right")
```

```python
# AREA PLOT The area between axis and line are commonly emphasized with colours
x=range(1,6)
n=[1,4,6,8,4]
# Area plot
plt.fill_between(x, n)
```

```python
## IRIS DATA SCATTER PLOT It shows the relationship between two variables.
# It displays the value of 2 sets of data on two-dimensions.
#Each dot represents an observation.
from sklearn import datasets
# Load Iris dataset
iris = datasets.load_iris()

# Extract the features and target variable
X = iris.data
y = iris.target

# Plot the scatter plot
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis')

# Set labels for the scatter plot
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.title('Iris Dataset: Sepal Length vs Sepal Width')

# Show the plot
plt.show()
```

```python
#SIMPLE BAR PLOT plot that presents categorical data with rectangular bars
plt.bar(x,n,color="pink")
```

```python
# grouped bar plot
x = np.arange(5)
y1 = [34, 56, 12, 89, 67]
y2 = [12, 56, 78, 45, 90]
width = 0.40

# plot data in grouped manner of bar type
plt.bar(x-0.2, y1, width,color="pink")
plt.bar(x+0.2, y2, width,color="yellow")
```

```python
#Stacked Bar Plot:it is used to show how a larger category is divided into smaller categories and what the relationship of each part has on the total amount
l = [4, 6, 2, 8, 9]
plt.bar(x, n,color="bisque")
plt.bar(x, m, bottom=n,color="lavender")
plt.bar(x, l, bottom=np.add(n, m), color='pink')

plt.show()
```

```python
#Histograms are used to visualize the distribution of a single continuous variable
plt.hist(x,color="pink")
```

```python
plt.hist(x, alpha=0.7, label='x',color="pink")
plt.hist(n, alpha=0.7, label='y',color="lavender")
plt.legend(loc='upper right')

plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of x and y')
```

```python
# circular statistical graphic, which is divided into slices to illustrate numerical proportion
animals = ['CATS', 'DOGS', 'RACCOONS',
        'PIGS', 'OTTTERS', 'HAMSTER']

likeability = [30, 70, 42, 35, 6, 41]

plt.pie(likeability, labels = animals)
plt.title("PIE CHART FOR ANIMALS")
plt.show()
```

```python
#joint plot between two variables, x and n
import seaborn as sns
sns.jointplot(x=x,y=n,kind="scatter")
```

```python
#One variable is represented on the horizontal axis, a second variable is represented on the vertical axis,
#third variable is represented by a colour gradient shows 3d in 2d
X,N=np.meshgrid(x,n)
Z = np.sin(np.sqrt(X**2 + N**2))
plt.contour(X,N,Z)
plt.colorbar()
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Contour Plot')

plt.show()
```
